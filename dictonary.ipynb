```python
student = {"name": "Alice", "age": 20, "course": "Python"}
print(student)
```

```python
d = {1: "Apple", 2: "Banana", 3: "Mango"}
print(d)
```

```python
student = {"name": "John", "age": 21}

print(student["name"])
print(student.get("age"))
```

```python
student = {"name": "John", "age": 21}

student["grade"] = "A"   # Add
student["age"] = 22      # Update

print(student)
```

```python
d = {"a": 1, "b": 2, "c": 3}

d.pop("b")
del d["c"]

print(d)
```

```python
d = {"a": 1, "b": 2}

for key in d:
    print(key, d[key])

# OR
for key, value in d.items():
    print(key, value)
```
