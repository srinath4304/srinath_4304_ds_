```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a | b)
print(a.union(b))
```

```python
a = {1, 2, 3}
b = {2, 3, 4}

print(a & b)
print(a.intersection(b))
```

```python
a = {1, 2, 3}
b = {2, 3, 4}

print(a - b)
```

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a ^ b)
```

```python
a = {1, 2}
b = {1, 2, 3, 4}

print(a.issubset(b))     # True
print(b.issuperset(a))   # True

```
